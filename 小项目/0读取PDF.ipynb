# PDF 词频统计

在英文文献过程中，主要是单词不认识，又不想因为要阅读一本书去背 3 个月的词典。顾采用程序方法，将高频词统计出来，在此基础上，查漏补缺，仅通过背少数词汇便可以达到阅读图书的目的。

1. 从 PDF 文件中解析字符串
2. 按照单词出现频率由高到低排列

3. 【暂定】调用翻译借口，给出单词的翻译内容
4. 输出词汇表


```python
import PyPDF2
```

```python
# 以二进制模式，打开文件
pdfFile0bj = open('./datafile/meetingminutes.pdf','rb')
```

```python
#  读取文件
pdfReader = PyPDF2.PdfFileReader(pdfFile0bj)
pdfReader.numPages# 查看文件页数
```

```python
# 将第一页内容返回回来
page0bj = pdfReader.getPage(0)

# 解析第一页内容的文本
page0bj.extractText()
```

```python
concent = ''
for i in range(pdfReader.numPages):
    page0bj = pdfReader.getPage(i)
    concent = concent + ' '+page0bj.extractText()
```
